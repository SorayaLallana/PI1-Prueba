Apertura - Lectura - convertir dataframe del archivo STEAM GAMES (Funciona ok)

```python
#1er paso: importo lo q voy a usar (pandas, gzip y json)

import pandas as pd
import gzip
import json

#2do paso: Nombre del archivo JSON comprimido y su variable
archivo_json_comprimido = "steam_games.json.gz"

#3er paso: creo una lista para almacenar los datos JSON q salen del archivo comprimido
datos = []

# 4to paso: Abro el archivo comprimido y cargo los datos JSON - gzipopen -> abre el archivo - rt -> abre archivo en txt - error ignore -> ignora los errores-
with gzip.open(archivo_json_comprimido, 'rt', encoding='utf-8', errors='ignore') as file:
    for line in file:
        try:
            # Convierte cada línea JSON en un diccionario de Python - strip -> borra espacion blancos   
            dato = json.loads(line.strip())
            datos.append(dato)
        except json.JSONDecodeError as e:
            print(f"Error al cargar JSON en la línea: {e}")

# 5to paso: Crea un DataFrame de Pandas a partir de los datos
df = pd.DataFrame(datos)


# 6to paso: se muestra el dataframe df especificamente los 1eros 13 datos
df.head(13)  
```

Código para abrir - leer y convertir en un df el archivo: user-reviews

```python
import pandas as pd
import gzip
import ast

def leer_archivo_comprimido():
    with gzip.open("user_reviews.json.gz", 'rt', encoding='utf-8', errors='ignore') as file:
        return [ast.literal_eval(line.strip())
                for line in file]

# lee el archivo comprimido a través de la función
misdatos = leer_archivo_comprimido()

# convierte los datos en un DataFrame de pandas
datosr = pd.DataFrame(misdatos)

datosr.head()
```

Desanidar steam games

```python
import pandas as pd
from pandas import json_normalize  # Importar la función json_normalize para desanidar

# Crear un DataFrame de ejemplo con una columna anidada
data = {'ID': [1, 2, 3],
        'columna_anidada': [{'nombre': 'Alice', 'edad': 28},
                            {'nombre': 'Bob', 'edad': 32},
                            {'nombre': 'Charlie', 'edad': 22}]}

df = pd.DataFrame(data)

# Desanidar la columna_anidada en un nuevo DataFrame
df_desanidado = json_normalize(df['columna_anidada'])

# Unir el DataFrame desanidado al DataFrame original usando la columna 'ID' como clave
df_final = pd.concat([df[['ID']], df_desanidado], axis=1)

# Ver el DataFrame final
print(df_final)
```

```python
import pandas as pd
from pandas import json_normalize
import gzip
import json

archivo_json_comprimido = "steam_games.json.gz"

datos = []
with gzip.open(archivo_json_comprimido, 'rt', encoding='utf-8', errors='ignore') as file:
    for line in file:
        try:
            dato = json.loads(line.strip())
            datos.append(dato)
        except json.JSONDecodeError as e:
            print(f"Error al cargar JSON en la línea: {e}")

df = pd.DataFrame(datos)

columna_anidada_desanidada = json_normalize(df['columna_anidada'])

df_final = pd.concat([df.drop('columna_anidada', axis=1), columna_anidada_desanidada], axis=1)
```
